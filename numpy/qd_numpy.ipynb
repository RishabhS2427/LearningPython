## Creating Arrays


> __NOTE__ : Numpy arrays are homogeneous in nature, i.e., they comprise one data type (integer, float, double, etc.) unlike lists.

```python
# Create arrays array of zeros
np.zeros(10, dtype='int')
```

```python
#creating a 3 row x 5 column matrix
np.ones((3,5), dtype=int)
```

```python
#creating a matrix with a predefined value
np.full((3,5),12.3)
```

```python
#create an array with set sequence
np.arange(0,20,2)
```

```python
#create an array of even space between the given range of values
np.linspace(0,2,10)
```

```python
#create a 3x3 array with mean 0 and standard deviation 1 in a given dimension
np.random.normal(0, 1, (3,3))
```

```python
#create an identity matrix 3X3
np.eye(3, dtype=int)
```

```python
#set a random seed
np.random.seed(0)
x1 = np.random.randint(10, size=6) #one dimension
print("x1 ndim:", x1.ndim)
print("x1 shape:", x1.shape)
print("x3 size: ", x1.size)
x2 = np.random.randint(10, size=(3,4)) #two dimension
print("x2 ndim:", x2.ndim)
print("x2 shape:", x2.shape)
print("x3 size: ", x2.size)

x3 = np.random.randint(10, size=(3,4,5)) #three dimension

print("x3 ndim:", x3.ndim)
print("x3 shape:", x3.shape)
print("x3 size: ", x3.size)
```

## Array Indexing
> __NOTE__ : The important thing to remember is that indexing in python starts at zero.

```python
x1 = np.array([4, 3, 4, 4, 8, 4])
x1
```

```python
#assess value to index zero
print(x1[0])
```

```python
#assess fifth value
print(x1[4])
```

```python
#get the second last value and last
print(x1[-1])
print(x1[-2])
```

```python
#in a multidimensional array, we need to specify row and column index
x2
```

```python
#1st row and 2nd column value
x2[2,3]
```

```python
#3rd row and last value from the 3rd column
x2[2,-1]
```

```python
#replace value at 0,0 index
x2[0,0] = 13
```

## Array Slicing
> __NOTE__ :  Access multiple or a range of elements from an array.

```python
x = np.arange(15)
print(x)
```

```python
#from start to 4th position
x[:5]
```

```python
#from 4th position to end
x[4:]
```

```python
#from 4th to 6th position
x[4:7]
```

```python
#return elements at even place
x[0::2]
```

```python
#return elements from first position step by two
x[1::2]
```

```python
#reverse the array
x[::-1]
```
