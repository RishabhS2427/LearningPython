##### Import libraries
##### create api obj and
##### get 30 days coin chart data by id : _bitcoin_

```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
cg = CoinGeckoAPI()
bitcoin_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency = 'usd', days=20,)
bit_df = pd.DataFrame.from_dict(bitcoin_data)
print(bitcoin_data)
```


##### Select only prices of bitcoins

```python
bitcoin_data['prices']
```

##### Convert into dataframe

```python

bit_df = pd.DataFrame(bitcoin_data['prices'], columns=['TimeStamp', 'price'])
bit_df
```

##### Convert timestamp into readable format

```python
bit_df['Date'] = pd.to_datetime(bit_df['TimeStamp'], unit='ms')
bit_df
```

### Ploting CandleSticks

```python
candlestick_data = bit_df.groupby(bit_df.Date.dt.date).agg({'price': ['min', 'max', 'first', 'last']})
candlestick_data
```

#### Use Plotly to render candles


```python
# download plotly and install
!pip3 install plotly
```

```python

import plotly.graph_objects as go
import plotly
fig = go.Figure(data=[go.Candlestick(x = candlestick_data.index,
                                           open= candlestick_data['price']['first'],
                                           high= candlestick_data['price']['max'],
                                           low= candlestick_data['price']['min'],
                                           close = candlestick_data['price']['last']
                                           )])
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title='Date', yaxis_title='Price $(USD)', title= 'Bitcoin Candlestick for 20 days')

plotly.offline.plot(fig, filename='./bitcoin_candlestick.html',)
```
